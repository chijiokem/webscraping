```python
import time
import pandas as pd

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys

from selenium import webdriver
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
# Add options as needed (e.g., headless mode, etc.)

# Set the path to the ChromeDriver executable
driver = webdriver.Chrome(options=chrome_options)

chrome_options = Options()


contract_url = []
all_data = []

driver.get('https://www.contractsfinder.service.gov.uk/Search/Results?page=1')


urls = driver.find_elements(By.XPATH, "//a[@class = 'govuk-link search-result-rwh break-word']")
for lnk in urls:
    all_links = lnk.get_attribute('href')
    contract_url.append(all_links)
for i in contract_url:
    driver.get(i)
    contract_details = driver.find_element(By.XPATH,"//div[@class = 'content-block']")
    contract_text = contract_details.text
    data = contract_text.split('\n')
    all_data.append(data)

dt = pd.DataFrame(all_data)
dt
```
